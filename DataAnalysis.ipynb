```python
def plotting(data) : 
  columns = data.columns
  fig = make_subplots(rows=4, cols=3, subplot_titles=columns)

  r = 1 
  c = 1 
  for clm in columns : 
    fig.add_trace(
        go.Histogram(x = data[clm], nbinsx = 15),
        row=r, col=c
    )
    c = c + 1
    if(c > 3):
      r = r + 1
      c = 1

  fig.update_layout(height=800, width=1000, title_text="Subplots")
  fig.update_traces(opacity = 0.75)
  fig.show()
```

```python
plotting(dframe)
```

#### Outlier Analysis

```python
# Calculate IQR
data=dframe
Q1 = data[['Battery mAh', 'Display(inches)', 'Price']].quantile(0.25)
Q3 = data[['Battery mAh', 'Display(inches)', 'Price']].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = ((data[['Battery mAh', 'Display(inches)', 'Price']] < (Q1 - 1.5 * IQR)) | 
            (data[['Battery mAh', 'Display(inches)', 'Price']] > (Q3 + 1.5 * IQR))).any(axis=1)

# Filter outliers
outliers_data = data[outliers]

print(outliers_data)
```

```python
# Scatter plot for Price vs Battery with outliers highlighted
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Battery mAh', y='Price', data=data, label='Normal Data')
sns.scatterplot(x='Battery mAh', y='Price', data=outliers_data, color='red', label='Outliers')
plt.legend()
plt.show()
```
